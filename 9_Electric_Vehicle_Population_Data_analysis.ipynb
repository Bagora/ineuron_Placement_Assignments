```python
# Get all the cars and their types that do not qualify for clean alternative fuel vehicles
non_qualify_cars = df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] != 'Clean Alternative Fuel Vehicle Eligible'][['Make', 'Model']]

# Count the number of cars for each make and sort in descending order
make_counts = non_qualify_cars['Make'].value_counts().sort_values(ascending=False)

print("Cars that do not qualify for clean alternative fuel vehicles:")
print(non_qualify_cars)

# Plot the distribution of cars that do not qualify for clean alternative fuel vehicles
plt.figure(figsize=(10, 6))
make_counts.plot(kind='bar')
plt.xlabel('Make')
plt.ylabel('Count')
plt.title('Cars that Do Not Qualify for Clean Alternative Fuel Vehicles (by Make)')
plt.show()
```

```python
# Get all TESLA cars with the model year and model type made in Bothell City
tesla_bothell_cars = df[(df['Make'] == 'TESLA') & (df['City'] == 'Bothell')][['Model Year', 'Model']]

# Count the number of cars for each model year and sort by the model year
model_year_counts = tesla_bothell_cars['Model Year'].value_counts().sort_index()

print("TESLA cars made in Bothell City")
print(tesla_bothell_cars)

# Plot the distribution of TESLA cars made in Bothell City by model year
plt.figure(figsize=(10, 6))
model_year_counts.plot(kind='bar')
plt.xlabel('Model Year')
plt.ylabel('Count')
plt.title('TESLA Cars Made in Bothell City (by Model Year)')
plt.show()
```

```python
# Get all the cars that have an electric range of more than 100 and were made after 2015
electric_cars = df[(df['Electric Range'] > 100) & (df['Model Year'] > 2015)]

# Count the number of cars for each make and sort in descending order
make_counts = electric_cars['Make'].value_counts().sort_values(ascending=False)

# Plot the distribution of cars with an electric range > 100 made after 2015
plt.figure(figsize=(10, 6))
make_counts.plot(kind='bar')
plt.xlabel('Make')
plt.ylabel('Count')
plt.title('Cars with Electric Range More Than 100 Made After 2015 (by Make)')
plt.show()
```

```python
# Draw plots to show the distribution between city and electric vehicle type
city_ev_type = df.groupby(['City', 'Electric Vehicle Type']).size().unstack()
city_ev_type.plot(kind='line')
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Distribution of Electric Vehicle Types in Cities')
plt.legend(title='Electric Vehicle Type')
plt.show()
```
