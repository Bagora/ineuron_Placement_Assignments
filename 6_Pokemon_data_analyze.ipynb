```python
# unique Pokémon types
unique_types = data["Type"].nunique()
print(f"Number of unique Pokémon types {unique_types}")
```

```python
# Get all Pokemons whose spawn rate is less than 5%
filtered_pokemon = data[data['Spawn Chance'] < 5]
print(filtered_pokemon)
```

```python
# Count the number of weaknesses for each Pokemon
data['Weaknesses Count'] = data['Weaknesses'].str.count(',') + 1

# Filter the DataFrame based on the number of weaknesses condition
filtered_pokemon = data[data['Weaknesses Count'] < 4]

# Print the filtered DataFrame
print(filtered_pokemon)
```

```python
# Get all Pokemons that have no multipliers at all
# Filter the DataFrame for Pokémon with no evolutions
filtered_pokemon = data[data['Next Evolution'].isnull() & data['Previous Evolution'].isnull()]

# Print the filtered DataFrame
print(filtered_pokemon)
```

```python
#Get all Pokemons that do not have more than 2 evolutions
filtered_pokemon = data[data['Next Evolution'].isnull() | data['Next Evolution'].str.count('>') <= 2]
print(filtered_pokemon)
```

```python
#To get all Pokemons whose spawn time is less than 300 seconds,
#we need to convert the spawn time from the format "minute:second" to seconds and then filter the DataFrame
# Convert spawn time to seconds
data['Spawn Time'] = data['Spawn Time'].fillna('00:00').apply(lambda x: int(x[:2]) * 60 + int(x[3:]))

# Filter the DataFrame for Pokemon with spawn time less than 300 seconds
filtered_pokemon = data[data['Spawn Time'] < 300]
print(filtered_pokemon)
```

```python
# Count the number of types for each Pokemon
data['Type Count'] = data['Type'].str.split(',').str.len()

# Filter the DataFrame for Pokemon with more than two types
filtered_pokemon = data[data['Type Count'] > 2]

# Print the filtered DataFrame
print(filtered_pokemon)
```

```python
import requests

def get_pokemon_with_multiple_types():
    pokemon_with_multiple_types = []

    # Fetching data for the first 151 Pokémon
    for pokemon_id in range(1, 152):
        url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_id}"
        response = requests.get(url)

        if response.status_code == 200:
            pokemon_data = response.json()

            # Checking if the Pokémon has more than two types
            if len(pokemon_data['types']) > 1:
                pokemon_name = pokemon_data['name']
                pokemon_with_multiple_types.append(pokemon_name)

    return pokemon_with_multiple_types

# Call the function to get the Pokémon with multiple types
pokemon_with_multiple_types = get_pokemon_with_multiple_types()

# Print the Pokémon with multiple types
for pokemon in pokemon_with_multiple_types:
    print(pokemon)
```
